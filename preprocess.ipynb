```python
# Import thư viện
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Đọc file csv
file1_path = 'D:\TTTN\CODE\Data Raw\data_2021.csv'
file2_path = 'D:\TTTN\CODE\Data Raw\data_2022.csv'
file3_path = 'D:\TTTN\CODE\Data Raw\data_2023.csv'

df1 = pd.read_csv(file1_path)
df2 = pd.read_csv(file2_path)
df3 = pd.read_csv(file3_path)

# Gộp ba file csv
df = pd.concat([df1, df2, df3], axis=0, ignore_index=True)

df.head(3)
```

```python
# Thông tin dữ liệu
df.info()
```

```python
# Tách cột 'Name' thành 2 cột 'Customer ID' và 'Customer Name'
df[['Customer ID', 'Customer Name']] = df['Name'].str.split(' ', n=1, expand=True)

df = df.drop('Name', axis=1)
df = df.drop('Customer Name', axis=1)
```

```python
# Xóa các cột không cần thiết
df = df.drop(['Sales Rep', 'Sales Employee', 'Terms', 'Amount Discount', 'UPC Code', 'Location'], axis=1)
```

```python
# Hiện thị giá trị null từng dòng
df.isnull().sum()
```

```python
# Xóa các dòng có giá trị null
df = df.dropna()
```

```python
# Kiểm tra sau khi xóa giá trị null
df.isnull().sum()
```

```python
# Tách cột 'Transaction Name' thành 2 cột 'Transaction Type' và 'Transaction Code'
df[['Transaction Type', 'Transaction Code']] = df['Transaction Name'].str.extract(r'([^\d]+)\s+(\S+)$')
df['Transaction Code'] = df['Transaction Code'].str.replace('#', '')

df = df.drop('Transaction Name', axis=1)
df = df.drop('Transaction Type', axis=1)
```

```python
# Mã hóa cột "Brand"
df['Brand'] = df['Brand'].replace('Vera : VERA BY CHIPU', 'VBC')
df['Brand'] = df['Brand'].replace('Jockey : JBU', 'JBU')
df['Brand'] = df['Brand'].replace('Jockey : JOC', 'JOC')
df['Brand'] = df['Brand'].replace('Vera : MIS', 'MIS')
df['Brand'] = df['Brand'].replace('Vera : VER', 'VER')
```

```python
# Định dạng lại cột 'Date Created'
df['Date Created'] = pd.to_datetime(df['Date Created'], format='%m/%d/%Y %H:%M')

# Chuyển đổi cột 'Quantity' sang kiểu int
df['Quantity'] = df['Quantity'].astype(int)

# Loại bỏ dấu phẩy và khoảng trắng từ chuỗi
df['Giá bán lẻ'] = df['Giá bán lẻ'].str.replace('(', '').str.replace(')', '').str.replace(',', '').str.strip()
df['Giá bán lẻ chưa thuế'] = df['Giá bán lẻ chưa thuế'].str.replace('(', '').str.replace(')', '').str.replace(',', '').str.strip()
df['Doanh thu chưa thuế'] = df['Doanh thu chưa thuế'].str.replace('(', '').str.replace(')', '').str.replace(',', '').str.strip()

# Đổi kiểu dữ liệu từ object sang float64
df['Giá bán lẻ'] = df['Giá bán lẻ'].astype(float)
df['Giá bán lẻ chưa thuế'] = df['Giá bán lẻ chưa thuế'].astype(float)
df['Doanh thu chưa thuế'] = df['Doanh thu chưa thuế'].astype(float)

# Thiết lập tùy chọn hiển thị để không sử dụng scientific notation
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
# Loại bỏ các giao dịch lỗi hoặc trả hàng
df = df[df["Quantity"] > 0]
df = df[df["Giá bán lẻ"] > 0]
```

```python
ds = ['Quantity', 'Giá bán lẻ', 'Giá bán lẻ chưa thuế', 'Doanh thu chưa thuế']
df[ds].describe()
```

```python
df['Date Created'] = pd.to_datetime(df['Date Created'])

# Tạo cột "Year month"
df['Year month'] = df['Date Created'].map(lambda x: str(x.year) + '-' + str(x.month))

# Tạo cột "Hour"
df['Hour'] = df['Date Created'].dt.hour

# Tạo cột "Day of week"
df['Day of week'] = df['Date Created'].dt.dayofweek

# Tạo cột "Day"
df['Day'] = df['Date Created'].dt.day_name()
```

```python
# Tạo cột "Weekend or weekday"
df['Weekend or weekday'] = df['Date Created'].apply(\
                                                     lambda x: 'weekend' if x.dayofweek // 5 == 1 else 'weekday')
df.head()
```
